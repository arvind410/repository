```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("data_clean.csv")
print(data)
```

```python
data.info()
```

```python
print(type(data))
print(data.shape)
print(data.size)
```

```python
#drop the duplicate columan (tem c ) and unnamed columan
# data1 = data.drop(['Unnamed: 0',"Temp C"], axis =1, inplace = True)
data1 = data.drop(['Unnamed: 0',"Temp C"], axis =1)
data1
```

```python
#convert the month column data into type to float data type

data1['Month']=pd.to_numeric(data['Month'],errors='coerce')
data1.info()
```

```python
data1[data1.duplicated(keep = False)]
```

```python
data1.drop_duplicates(keep='first', inplace = True)
data1
```

```python
#change columan names (renames the columns)

data1.rename({'Solar.R': 'Solar'}, axis=1, inplace = True)
data1
```

```python
data1.isnull().sum()
```

```python
cols = data1.columns
colors = ['red', 'white']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar = True)
```

```python
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean()
print("Median of Ozone: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
data1['Ozone'] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
median_solar = data1["Solar"].median()
mean_solar = data1["Solar"].mean()
print("Median of Solar: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
data1['Solar'] = data1['Solar'].fillna(median_solar)
data1.isnull().sum()
```

```python
data1['Solar'] = data1['Solar'].fillna(mean_ozone)
data1.isnull().sum()
```

```python
#prit the data1 5 rows
data1.head()
```

```python
#find the mode values of categorical column (weather)

print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python
data1["Weather"] = data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
mode_month = data1["Month"].mode()[0]
data1["Month"] = data1["Month"].fillna(mode_month)
data1.isnull().sum()
```
